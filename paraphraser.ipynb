```python
# import libraries
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings('ignore')
```

```python
# for reproducibility
def random_state(seed):
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        
random_state(1234)
```

```python
# init models(make sure you init only once if you integrate this to other code)
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
# phrases = ['Can you tell me which is the best city in the world?','Where can I find the best momos?']
# phrases = 'I’d love to see some of your creations and so please feel free to post them in the comment section. Happy creation!'
phrases = str(input("Please enter the text to be paraphrased: "))
```

```python
ip_para = list(map(str, phrases.split('.')))
print(ip_para)
```

```python
op_para = ""
for phrase in ip_para:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase.strip())
  # for para_phrase in para_phrases:
  #   # print(para_phrase)
  #   op_para=op_para+str(para_phrase[0])+" "
  op_para=op_para+str(para_phrases[0][0]).capitalize()
  last_char = str(para_phrases[0][0])[-1]
  print(last_char)
  if last_char=="?" or last_char=="!":
    op_para=op_para+" "
  else:
    op_para=op_para+". "
print(op_para)
```
